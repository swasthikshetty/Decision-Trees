```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fraud = pd.read_csv("C:\\Users\\Swathi\\Desktop\\Excelr Classes\\Assignments\\12.Decision Trees\\Fraud_check.csv")
fraud_check = fraud
fraud_check.head()
```

```python
fraud_check=fraud_check.rename(columns={'Marital.Status': 'Marital_Status', 'Taxable.Income':'Taxable_Income', 
                          'City.Population':'City_Population', 'Work.Experience':'Work_Experience'})
fraud_check.head()
```

```python
fraud_check['income_range'] = np.where(fraud_check['Taxable_Income']>=30000, 'Good', 'Risky')
fraud_check.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label= LabelEncoder()
fraud_check1 = fraud_check[['Undergrad', 'Marital_Status','Urban']].apply(label.fit_transform)
fraud_check1 = fraud_check1.rename(columns={'Undergrad':'Undergrad_num', 'Marital_Status':'Marital_num','Urban':'Urban_num'})
fraud_check1.head()
```

```python
fraud_check = pd.concat([fraud_check,fraud_check1], axis=1, sort=False)
fraud_check.head()
```

```python
fraud_check =fraud_check.drop(columns={'Undergrad','Marital_Status','Urban','Taxable_Income'},axis=1)
fraud_check.head()
```

```python
fraud_check = fraud_check[['City_Population','Work_Experience','Undergrad_num','Marital_num','Urban_num','income_range']]
fraud_check.head()
```

```python
colnames=list(fraud_check.columns)
colnames
```

```python
predictors = colnames[:5]
target=colnames[5]
print("predictors:\n",predictors)
print("target:",target)
```

```python
fraud_check.isnull().any()
```

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(fraud_check,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

model1=DecisionTreeClassifier(criterion='entropy')
model1.fit(train[predictors],train[target])

pred = model1.predict(test[predictors])
print(pred)
pd.Series(pred).value_counts()
```

```python
print(pd.crosstab(test[target],pred))
```

```python
print(np.mean(pred==test.income_range))
```
