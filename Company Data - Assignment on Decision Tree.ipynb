```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
companydata = pd.read_csv("C:\\Users\\Swathi\\Desktop\\Excelr Classes\\Assignments\\12.Decision Trees\\Company_Data.csv")
company_data=companydata
company_data.head()
```

```python
company_data['Sales_Range'] = pd.cut(company_data.Sales,bins=[0,3,6,10,14,20],
                                     labels=['Poor','Below_Avg','Avg','Above_Avg','Outstanding'],include_lowest=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label= LabelEncoder()
company_data1 = company_data[['ShelveLoc', 'Urban','US']].apply(label.fit_transform)
company_data1 = company_data1.rename(columns={'ShelveLoc':'ShelveLoc_num', 'Urban':'Urban_num','US':'US_num'})
company_data1.head()
```

```python
company_data = pd.concat([company_data,company_data1], axis=1, sort=False)
company_data.head()
```

```python
company_data =company_data.drop(columns={'Sales','ShelveLoc','Urban','US'},axis=1)

company_data = company_data[["CompPrice","Income","Advertising","Population","Price","Age",
                             "Education","ShelveLoc_num","Urban_num","US_num","Sales_Range"]]


company_data.head()
```

```python
colnames=list(company_data.columns)
colnames
```

```python
predictors = colnames[:10]
target=colnames[10]
print("predictors:\n",predictors)
print("target:",target)
```

```python
company_data.isnull().any()
```

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(company_data,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

model1=DecisionTreeClassifier(criterion='entropy')
model1.fit(train[predictors],train[target])

pred = model1.predict(test[predictors])
print(pred)
pd.Series(pred).value_counts()
```

```python
print(pd.crosstab(test[target],pred))
```

```python
print(np.mean(pred==test.Sales_Range))
```
